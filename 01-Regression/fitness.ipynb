Create a Regression model that would plot the relationship between the waistline and how many situps are accomplished.

```python
from sklearn.datasets import load_linnerud
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
```

**Load the Dataset**

```python
# Load dataset
data = load_linnerud()

# Convert to DataFrame for better visualization
X = pd.DataFrame(data.data, columns=data.feature_names) # Exercise data
y = pd.DataFrame(data.target, columns=data.target_names) # Physiological data

print(X.head())
print(y.head())
```

**To analyze the relationship between Situps and Waistline, we need to use Situps (from X) as the independent variable and Waist (from y) as the dependent variable.**

```python
# Define input (Situps) and output (Waist)
X_situps = X[['Situps']]  # Situps column as a 2D array
y_waist = y[['Waist']]    # Waist column as a 2D array
```

**Split Dataset**

```python
# Split the data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_situps, y_waist, test_size=0.33)

# 2/3 of the data is used for training
# 1/3 of the data is used for testing
```

**Train the model**

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

**Predict waistline values**

```python
# Predict using our test values
y_pred = model.predict(X_test)
```

**Show the plot**

```python
# Plot the regression line
plt.scatter(X_test, y_test, color='black', label='Test Data')

# Plot the predictions
plt.plot(X_test, y_pred, color='blue', linewidth=2, label='Regression Line')

# Labels and title
plt.xlabel('Number of Situps')
plt.ylabel('Waistline (cm)')
plt.title('Relationship Between Situps and Waistline')

# Show the plot
plt.legend()
plt.show()
```

## Regression for other relationships

**Chins vs. Weight**

```python
X_chins = X[['Chins']]
y_weight = y[['Weight']]

model.fit(X_chins, y_weight)
y_pred = model.predict(X_chins)

plt.scatter(X_chins, y_weight, color='black')
plt.plot(X_chins, y_pred, color='blue', linewidth=2)
plt.xlabel('Number of Chin-ups')
plt.ylabel('Weight (kg)')
plt.title('Relationship Between Chin-ups and Weight')
plt.show()
```

**Jumps vs. Pulse**

```python
X_jumps = X[['Jumps']]
y_pulse = y[['Pulse']]

model.fit(X_jumps, y_pulse)
y_pred = model.predict(X_jumps)

plt.scatter(X_jumps, y_pulse, color='black')
plt.plot(X_jumps, y_pred, color='blue', linewidth=2)
plt.xlabel('Number of Jumps')
plt.ylabel('Pulse Rate (bpm)')
plt.title('Relationship Between Jumps and Pulse Rate')
plt.show()
```
